```python
def add_and_sum(x,y):
    added = x + y
    summed = added.sum(axis=1)
    return summed
def call_function():
    x = randn(1000,1000)
    y = randn(1000,1000)
    return add_and_sum(x,y)
```

```python
x = randn(1000,1000)
y = randn(1000,1000)
%prun add_and_sum(x,y)
```

```python
%lprun -f add_and_sum add_and_sum(x,y)
```

```python
%lprun -f add_and_sum -f call_function call_function()
```
