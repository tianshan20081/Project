```python
def get_counts(sequences):
    counts = {}
    for i in sequences:
        if i not in counts :
            counts[i] = 1
        else :
            counts[i] += 1
    return counts
counts = get_counts(time_zones)
counts['America/New_York']
```

```python
len(time_zones)
```

```python
from collections import defaultdict
def get_counts2(sequences):
    counts = defaultdict(int)
    for x in sequences :
        counts[x] += 1
    return counts
cts = get_counts2(time_zones)
cts['America/New_York']
```

```python
# 获取前十位 时区 和数量
def top_counts(counts_dict,top = 10):
    value_key_pairs = [(count,tz) for tz,count in counts_dict.items()]
#     for t,c in counts_dict.items():
#         print(t,c)
    value_key_pairs.sort()
#     len(value_key_pairs)
    return value_key_pairs[-top:]
top_counts(cts)
# len(cts.items())
# t,c = cts.items()[0]
```

```python
cframe = frame[frame.a.notnull()]
cframe[:10]
```

```python
operating_system = np.where(cframe['a'].str.contains('Windows'),'Windows','Not Windows')
operating_system[:10]
```

```python
by_tz_os = cframe.groupby(['tz',operating_system])
agg_counts = by_tz_os.size().unstack().fillna(0)
agg_counts[:10]
```

```python
indexer = agg_counts.sum(1).argsort()
indexer[:10]
```

```python
count_subset = agg_counts.take(indexer)[-10:]
count_subset
```

```python
count_subset.plot(kind='barh',stacked=True)
```

```python
normed_subset = count_subset.div(count_subset.sum(1),axis=0)
normed_subset.plot(kind='barh',stacked=True)
```
