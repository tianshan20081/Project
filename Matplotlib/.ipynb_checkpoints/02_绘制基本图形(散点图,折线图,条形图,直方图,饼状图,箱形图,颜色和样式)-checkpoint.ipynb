```python
import codecs
filecp = codecs.open('000003.csv', encoding ='ISO-8859-1')
# print(datas)
open,close = np.loadtxt(filecp,delimiter=',',skiprows=1,usecols=(4,3),unpack=True)
change = close - open
# change
yestday = change[:-1]
today = change[1:]
plt.scatter(yestday,today,c='r',marker='o')
plt.show()
```

```python
filecp = codecs.open('000003.csv', encoding ='ISO-8859-1')
m,open = np.loadtxt(filecp,delimiter=',',skiprows=1,usecols=(6,4),unpack=True)
change = m - open
yestday = change[:-1]
today = change[1:]
plt.scatter(yestday,today)
plt.show()
```

```python
import matplotlib.dates as mdates
def bytedate2num(fmt):
    def converter(b):
#         print('ascii')
        return mdates.strpdate2num(fmt)(b.decode('ascii'))
    return converter

date_converter = bytedate2num("%Y-%m-%d")

filecp = codecs.open('000003.csv',encoding='ISO-8859-1')
date,open,close = np.loadtxt(filecp,delimiter=',',converters={0: date_converter},
                            skiprows=1,usecols=(0,6,3),unpack=True)
plt.plot(date,open)
plt.show()
```

```python
plt.plot_date(date,open,'-')
```

```python
plt.plot_date(date,open,linestyle='-',color='r')
```

```python
plt.plot_date(date,open,linestyle='--',color='b',marker='<')
```

```python
'''
03 条形图 http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.bar 

bar(left, height, width=0.8, bottom=None, hold=None, data=None, **kwargs)

'''
# bar

y = [20,10,15,5,6]
index = np.arange(5)
print(index)
plt.bar(left=index,height=y)
plt.show()
```

```python
plt.bar(left=index,height=y,color='r',width=0.5)
```

```python
plt.bar(left=0,bottom=index,height=0.5,color='r',width=y,orientation='horizontal')
plt.show()
```

```python
plt.barh(left=0,bottom=index,width=y)
plt.show()
```

```python
index = np.arange(4)
sales_BJ = [20,30,50,80]
sales_SH = [15,31,55,75]
bar_width = 0.3
plt.bar(index,sales_BJ,bar_width,color='b')
plt.bar(index+bar_width,sales_SH,bar_width,color='r')
```

```python
plt.bar(index,sales_BJ,bar_width,color='b')
plt.bar(index,sales_SH,bar_width,color='r',bottom=sales_BJ)
```

```python
'''
04 直方图
plt.hist()
'''
```

```python
num = 100
sigma = 20
x = num + sigma * np.random.randn(2000)
plt.hist(x,bins=10,color='r',normed=True)
plt.show()
```

```python
plt.hist(x,bins=50,color='g',normed=True)
plt.show()
```

```python
x = np.random.randn(10000)+2
y = np.random.randn(10000)+3
plt.hist2d(x,y,bins=20)
plt.show()
```

```python
'''
05 饼状图
plt.pie()
'''
labels = list('ABCD')
fraces = [15,30,40,55]

plt.pie(x=fraces,labels=labels)
plt.show()
```

```python
plt.axes(aspect=1)
plt.pie(x=fraces,labels=labels)
plt.show()
```

```python
plt.axes(aspect=1)
explode = [0,0,0.5,0.1]
plt.pie(x=fraces,labels=labels,autopct='%.0f%%',explode=explode)
plt.show()
```

```python
plt.axes(aspect=1)
explode = [0,0,0.5,0.1]
plt.pie(x=fraces,labels=labels,autopct='%.0f%%',explode=explode,shadow=True)
plt.show()
```

```python
'''
06 箱形图
plt.boxplot()
'''
data = np.random.normal(size=1000,loc=0,scale=1)
plt.boxplot(data,sym='o',whis=1.5)
plt.show()
```

```python
data = np.random.normal(size=(1000,4),loc=0,scale=1.5)
labels = list('ABCD')
plt.boxplot(data,whis=1,labels=labels)
plt.show()
```

```python
'''
07 颜色和样式
'''
y = np.arange(1,5)
plt.plot(y,color='g')
plt.plot(y+1,color='0.5')
plt.plot(y+2,color='#FF00FF')
plt.plot(y+3,color=(0.1,0.2,0.5))
plt.show()
```

```python
y = np.arange(1,5)
plt.plot(y,color='g',marker='o')
plt.plot(y+1,'D')
plt.plot(y+2,'*')
plt.plot(y+3,'p')
plt.show()
```

```python
y = np.arange(1,5)
plt.plot(y,color='g',marker='o')
plt.plot(y+1,marker='D')
plt.plot(y+2,'*')
plt.plot(y+3,'p')
plt.show()
```

```python
y = np.arange(1,5)
plt.plot(y,'--',color='g')
plt.plot(y+1,'-')
plt.plot(y+2,':')
plt.plot(y+3,'-.')
plt.show()
```

```python
y = np.arange(1,5)
plt.plot(y,'cx--',color='g')
plt.plot(y+1,'kp-')
plt.plot(y+2,':')
plt.plot(y+3,'mo-.')
plt.show()
```
