```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Polygon
%matplotlib inline
```

```python
def func(x):
    return -(x-2)*(x-8) + 40

x = np.linspace(0,10)

y = func(x)
fig ,ax = plt.subplots()
plt.plot(x,y,'r',linewidth=2)
```

```python
x = np.linspace(0,10)

y = func(x)
fig ,ax = plt.subplots()
plt.plot(x,y,'r',linewidth=2)

a = 2 
b = 9

ix = np.linspace(a,b)
iy = func(ix)
ixy = zip(ix,iy)

verts = [(a,0)] + list(ixy) + [(b,0)]

ploy = Polygon(verts,facecolor='0.9',edgecolor='0.1')
# ploy.set_color('g')

ax.add_patch(ploy)
ax.set_xticks([a,b])
ax.set_yticks([])

ax.set_xticklabels(['$a$','$b$'])

plt.figtext(0.9,0.05,'$x$')
plt.figtext(0.1,0.9,'$y$')

x_math = ( a + b )*0.2
y_math = 35
plt.text(x_math,y_math,r'$ \int_a^b (-(x-2)*(x-8) + 40)dx$',fontsize=15)

plt.ylim(25)
```
